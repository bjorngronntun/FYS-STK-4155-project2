```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Load breast cancer
data = load_breast_cancer()
X, y = data.data, data.target
```

```python
# Load credit card data
df = pd.read_excel('../data/raw/default of credit card clients.xls')
df.head()
```

Do some relabeling

```python
df_names = list(df.iloc[0])
df_names[-1] = 'Y'
df.columns = df_names
df.head()
```

Drop first row

```python
df = df[1:]
df.head()
```

```python
df.info()
```

```python
# Looks good. Let us try some extremely naive predicting
all_data = np.array(df)

# Of course, some of the columns should be treated as categorical. But now we pretend they are numerical.
X = all_data[:, :-1].astype('float')
y = all_data[:, -1].astype('int')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Logistic regression!
lr = LogisticRegression(solver='lbfgs', max_iter=200)
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
# Well.
```
